**17)** [P] Write a function called `IQR_outlier_limits` that takes a dataframe as input, and computes the minimum and maximum outlier thresholds for each variable (numeric only assumed), stored in a data frame that is returned as a result.

````python
# ANSWER:

def IQR_outlier_limits(df):
    """ Compute the outlier limits based on the 1.5 * IQR rule
        Input: A DataFrame
        Output: a DataFrame of the minimum and maximum outlier thresholds for all 
                numeric variables
    """
    stats = df.describe(exclude = 'category')
    
    IQR = stats.loc['75%'] - stats.loc['25%']
    
    min_out = stats.loc['25%'] - 1.5 * IQR
    max_out = stats.loc['75%'] + 1.5 * IQR
    
    outlier_thresholds = pd.DataFrame([min_out, max_out], index = ['min_out', 'max_out'], columns = ["age", "numvisit"])
    
    return outlier_thresholds
    






````

# Part III - Discretization
---

As we learn about doing EDA and prepping our data, we're learning that visualizing our data to get a good understanding of our variables is important. We're also learning that we can make a variety of transformations with our variables in ways that let us do more meaningful visualizations (and also, as we'll soon understand, modeling).

One common transformation is called **discretization**. It's the process of converting a numeric variable into a categorical variable, more specifically, usually an ordinal.

A common way of referring to categorical variables are **factors**. Factors are explanatory categorical variables in a controlled experiment. They typically only take on a small number of values, called **levels**.

---
**42)**	[P] Create an additional variable in df_badhealth called `age_fac1` that represents a factor with 3 levels, "young", "mid", and "older". You should discretize according to equal width bins over the distribution of the age variable.



````python
bin_edges = np.histogram(badHealth['age'], bins = 3, range = [badHealth['age'].min(), badHealth['age'].max()])[1]
badHealth["age_fac1"] = pd.cut(badHealth['age'], bin_edges, labels = ["young", "mid", "older"])


````
